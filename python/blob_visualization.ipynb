```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH_TO_DATA = "../pfmdData/blob_data_files/"
data = pd.read_parquet(PATH_TO_DATA+ "/" + 'blob_log.parquet', engine='pyarrow')
```

```python
data.head(20)
```

```python
plt.figure(1)
max_size = 50
min_size = 25
plt.plot(data["x"][data["cam"]==1][data["pixel_count"]<max_size][data["pixel_count"]>min_size], \
         data["y"][data["cam"]==1][data["pixel_count"]<max_size][data["pixel_count"]>min_size], '.', label="cam 1")
plt.plot(data["x"][data["cam"]==0][data["pixel_count"]<max_size][data["pixel_count"]>min_size], \
         data["y"][data["cam"]==0][data["pixel_count"]<max_size][data["pixel_count"]>min_size], 'r.', label="cam 0")
plt.title("z_up_10 track")
plt.legend()
plt.gca().set_aspect('equal')
plt.show()
```
